```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
```

```python
tasks = [
    ("Research Work", "August", "October"),
    ("Data Collection", "September", "September"),
    ("Preprocessing & Feature Selection", "September", "January"),
    ("Model Selection", "August", "January"),
    ("Training and Testing", "October", "April"),
    ("Debugging", "February", "April"),
    ("Model Optimization", "December", "May"),
    ("Data Optimization", "November", "June"),
    ("Documentation", "December", "June"),
]

months = [
    "August", "September", "October", "November", "December", 
    "January", "February", "March", "April", "May", "June"
]
month_dates = {month: datetime.datetime(2025, i + 8 if i < 5 else i - 4, 1) for i, month in enumerate(months)}

task_bars = [(task[0], month_dates[task[1]], month_dates[task[2]]) for task in tasks]

# Plot Gantt chart
fig, ax = plt.subplots(figsize=(12, 6))

for i, (task, start, end) in enumerate(task_bars):
    ax.barh(task, (end - start).days, left=start, color=plt.cm.Paired(i / len(tasks)), align='center')

# Formatting
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.invert_yaxis()  # Invert y-axis for a top-to-bottom Gantt chart
ax.set_xlabel("Timeline")
ax.set_ylabel("Tasks")
ax.set_title("Project Timeline Gantt Chart")
plt.grid(axis="x", linestyle="--", alpha=0.5)

# Show the chart
plt.show()
```
